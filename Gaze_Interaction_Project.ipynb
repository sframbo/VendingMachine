# Features implemented:

*   **Item selection using dwell interaction**: Hovering over an item triggers a loading icon which after a given time will activate a purchase confirmation prompt. This is done by assigning an Snack object to each AOI location as well implementing a dwell counter which counts down when a gaze enters an AOI. If the gaze leaves the AOI, the dwell counter is reset back to its specified threshold and no other action is performed. Otherwise, a confirm-purchase prompt is opened and the dwell counter is reset. 
*   **Purchase confirmation using head movement (nodding)**: You will be prompted to nod to confirm purchase. If no nod movement is detected within a given time frame, the process is cancelled. This is done by implementing a landmark detector using OpenCV and then extracting the "chin" (the point with the lowest y-position) from the landmark. When a change of y-position higher than a specified threshold is detected, a nod is detected. (landmark detection based on https://medium.com/analytics-vidhya/facial-landmarks-and-face-detection-in-python-with-opencv-73979391f30e) 
*   **Transaction process using Orbits**: When a nod action is detected during the prompt, a white circle will begin hovering around the screen. Follow the circle with your eyes within a given time frame to finish transaction. To do this, the target position and the corresponding gaze position is stored within a given time frame. The Pearson's correlation coefficient of the x- and y-positions between the target and the gaze is calculated via a built-in function in NumPy. The minimum of which is compared to a minimum threshold. (Algorithm based on: https://dl.acm.org/doi/10.1145/2807442.2807499)

This eyetracker relies on keeping your head still which is difficult to do in real situations. To make this work more properly for longer periods of time, perhaps some recalibrations should be done in-game --not unlike the (self-)recalibration process implemented in Webgazer (https://www.researchgate.net/publication/305375498_WebGazer_Scalable_Webcam_Eye_Tracking_Using_User_Interactions). We tried to implement something similar by updating the gaze-cursor relation at every click and then recalibrating but it didn't seem to improve the accuracy much. 

Another issue is choosing the proper eye/head movements to interact with the GUI. We purposely chose not to use the blink detection because we didn't want to use an involuntary movement but the nodding action was also just as problematic due to potential neck strain or vertigo (not to mention moving the head and therefore messing with the accuracy).  

As mentioned in the lecture before, the smooth pursuit method seems a bit cumbersome when you could easily finish the same task with an alternative eye/head movement for less time.
